```python
import pandas as pd

df = pd.read_csv('/content/cleanedmushroomdata.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df['class'] = df['class'].astype(int)
```

```python
df.head(3)
```

```python
df.head(5)
```

```python
from sklearn import set_config
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

set_config(transform_output='pandas')
```

```python
columns = df.columns.tolist()
columns.remove('class')
```

```python
columns
```

```python
x = df[columns]
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
trf = ColumnTransformer([('scaler', StandardScaler(), columns)], remainder='passthrough' )
```

```python
x_train_trf = trf.fit_transform(x_train)
x_test_trf = trf.transform(x_test)
```

```python
x_train_trf
```

```python
lr = LogisticRegression()
lr.fit(x_train_trf, y_train)
```

```python
pred = lr.predict(x_test_trf)
```

```python
score = accuracy_score(y_test, pred)*100
score
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train_trf, y_train)
```

```python
pred_new = dtc.predict(x_test_trf)
```

```python
new_score = accuracy_score(y_test, pred_new)*100
new_score
```
